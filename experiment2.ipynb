```python
import pandas as pd
import numpy as np
data={
    'Firstname':['aryan','rohan','riya','yash','siddini'],
    'Lastname':['singh','agarwal','shah','batia','khan'],
    'Type':['full_time employee','return','full_time employee','part_time employee','full_time employee'],
    'Department':['administration','techncal','administration','technical','technical'],
    'yoe':[2,3,5,7,6],
    'salary':[20000,5000,10000,2000,30000]
}
df=pd.DataFrame(data)
print(df)
table=pd.pivot_table(df,index=['Type','Department'],values='salary',aggfunc=['mean'])
print(table)
table=pd.pivot_table(df,index=['Type','Department'],values='salary',aggfunc=['sum','mean'])
print(table)
```
